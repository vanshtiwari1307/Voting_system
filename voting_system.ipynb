# Login Page

```python
def loginpage():
    
    root2 = Tk()
    root2.geometry("1080x1980")
    root2.configure(bg="lightblue")
    
    frame = Frame(root2,bg="pink", width=300, height=200)
    frame.place(x=200,y=200)
    def login():
        adharno = ade.get()
        pas = pae.get()
        cur = conn.cursor()
        sql = """SELECT * FROM userinfo WHERE adhar = %s AND password = %s"""
        cur.execute(sql, (adharno, pas))
        result = cur.fetchone()
        if result:
            messagebox.showinfo("Login Success", f"Welcome, {result[1]}!")  # Adjust index as per table
            root2.destroy()
            voting()  #............................................................................................................
        else:
            messagebox.showerror("Login Failed", "Invalid Adhar No. or Password")
        print("Login Successfully")
    
    
    he = Label(frame,text="Login Page",font=("Courier New",30,"bold"),relief = SUNKEN,padx=20,pady=10,bg="black",fg="lightblue")
    he.grid(row=0,column=0,columnspan = 2,pady=30)
    ad = Label(frame,text="Adhar No.:",font=("Courier New",20,"bold"),relief = SUNKEN,padx=10,pady=5)
    ad.grid(row=1,column=0,pady=10,padx=10)
    pa = Label(frame,text="PassWord:",font=("Courier New",20,"bold"),relief = SUNKEN,padx=10,pady=5)
    pa.grid(row=2,column=0,pady=10)
    
    ade = Entry(frame,width=15,font=("Courier New",25,"bold"),relief = SUNKEN)
    ade.grid(row=1,column=1,padx=10)
    pae = Entry(frame,font=("Courier New",25,"bold"),relief = SUNKEN,width=15,show="*")
    pae.grid(row=2,column=1)
    
    sub = Button(frame,text="Submit",font=("Courier New",30,"bold"),padx=2,pady=2,command=login)
    sub.grid(row=3,column=0,columnspan=2,pady=20)
    
    
    root2.mainloop()
```
